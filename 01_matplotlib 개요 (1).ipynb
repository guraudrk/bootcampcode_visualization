# 데이터 시각화의 개념 및 필요성

## 데이터 시각화
- 데이터의 분석 결과를 쉽게 이해하고 판단할 수 있도록 데이터를 시각적으로 표현하여 전달하기 위한 과정을 말한다.

## 필요성
- 많은 양의 데이터를 한눈에 파악할 수 있다. 시각적 요소를 활용해 데이터를 요약한다.
- 데이터 분석에 대한 전문적 지식이 없더라도 누구나 해당 데이터를 인지하고 활용할 수있게한다.
    - 사람이 감각기관을 통해 획득하는 정보의 80%는 시각에 의한 것이다.(이게 은근히 중요하다.)
    - 시각적인 입력은 다른 어떤 방법보다 빠르고 쉽게 사람이 이해할 수 있다.
    - 시각화 형태, 요소, 위치, 색등을 기반으로 패턴이나 다양한 인사이트(의미있는 정보)를 도출 및 표현할 수 있다.
- 단순한 데이터 요약, 통계분석 결과보다 정확한 데이터 분석결과를 도출할 수있다.
    - 데이터의 패턴이나 추세, 데이터간 비교등 표(Table)로 쉽게 인식 할 수 없는 것들을 시각적으로 쉽게 확인할 수 있다.

## matplotlib 그래프 구성요소(axis를 subplot이라고도 한다.)
- 상당히 많은 요소가 있는데, 이건 솔직히.... 하면서 아는게 가장 좋다.


```python
#아래의 것들은....뭐랄까 그냥 해보면서 눈에 익히면 좋다.
```

- **figure**
    - 전체 그래프가 위치할 기본 틀
    - 하나의 figure에 여러개의 그래프를 그릴 수 있다.
    - https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.figure.html
- **axes(subplot)**
    - figure안에 axes를 만들어 놓고 활동을 하는 로직이다.
    - figure내에 한개의 그래프를 그리기 위한 공간
        - figure에 한개 이상의 axes(subplot)로 구성해서 각 axes에 그래프를 그린다.
    - https://matplotlib.org/stable/api/axes_api.html
- **axis** 
    - 축 (x축, y축)은 값들을 위치시키는 선을 말한다.
    - axis label (x, y) : 축의 레이블(설명)
    - axis의 수는 곧 차원을 의미한다. 2이면 2차원이고 3이면 3차원이다.
- **ticks** : 축의 값을 알려주는 눈금
    - Major tick
    - Minor tick
- **title** : 플롯 제목   
- **legend (범례)**
    - 하나의 axes내에 여러 그래프를 그린 경우 그것에 대한 설명

```python
 %matplotlib inline
#%matplotlib qt 
#이거를 활성화시키면 다른 창에 그래프가 뜬다.


#그래프들의 예시. 자세한 것은 지금은 몰라도 된다.
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure(figsize=(15,7), facecolor='gray') #facecolor: figure의 배경색////figure는 배경을 의미한다.
axes1 = fig.add_subplot(1,2,1)
axes2 = fig.add_subplot(1,2,2)

axes1.plot([1,2,3,4,5], [10,20,30,40,50], label='line1')
axes1.plot([1,2,3,4,5], [50,40,30,20,10], label='line2')
axes2.scatter(np.random.randint(100, size=50), np.random.randint(100, 200, size=50), color='r')

fig.suptitle('Example of Plot', size=25, color='blue') #size: 폰트크기, color: 글자색
axes1.set_title("PLOT 1", size=20)
axes2.set_title("Plot 2", size=20)

axes1.set_xlabel("X축", size=15)
axes1.set_ylabel("Y축", size=15)
axes2.set_xlabel("가격1", size=15)
axes2.set_ylabel('가격2', size=15)

axes1.legend()
axes1.grid(True)
plt.show()
```

## pyplot 모듈을 이용해 그리기
1. pyplot 모듈이 그래프 그리는 함수와 Axes(Subplot) 설정 관련 함수를 제공

```python
#그래프에 사용할 데이터
x=[1,2,3,4,5] #x축
y=[10,20,30,40,50] #y축


```

```python
import matplotlib.pyplot as plt #plt로 alias를 주는 것을 기억하자.

#그래프를 그리기! plt 모듈의 함수이용.
plt.plot(x,y) #line plot(선그래프)

#그래프에 대한 설정을 해주자.

#1.제목설정
plt.title("lineplot 예제") #제목

#2.label 설정
plt.xlabel("X값") #x,y 축의 label 설정
plt.ylabel("Y값")


#3.grid-----그래프의 선을 그어주는 메소드.

plt.grid(True)



#4.눈금(ticks) 설정 ---- 눈금을 어디에 붙일 것인지?
#0부터 5까지 였는데, 0부터 10까지로 늘어났다.
#labels는 눈금의 이름을 바꿀 수 있다. list 형태와 결합해서 쓸 수도 있음.
plt.xticks(ticks=range(0,11),labels=list("ABCDEFGHIJK"))

#y축 눈금.
#plt.yticks(ticks=range(10,51,5)) 



# 그래프를 출력한다. 출력하는 메소드는 show 이다.
#눈금 설정을 안하면, 알아서 만들어준다.
#위의 과정들을 미리 하지 않는다. plt.show()을 실행을 할 때, 한번에 다 실행을 한다. 
#윗줄과 같은 것을 지연 랜더링이라고 한다.
plt.show()
```

2. 하나의 figure에 여러 그래프 그리기
    - plt.subplot(row,columns,num)

```python
#이 코드는 matplotlib으로 그래프를 그리는 데에 가장 기초가 되는 코드이다. 숙지!!


import matplotlib.pyplot as plt


#0.크기 설정
plt.figure(figsize=(10,5)) #figure의 크기를 설정.
#10은 가로길이, 5는 세로 길이. 단위는 인치이다.


#1.자, 그래프를 그리기 전에...
#axes == subplot(그래프를 그리는 영역).figure에 포함이 된다.
plt.subplot(1,2,1) #figure을 1행 2열의 형태로 나누겠다는 것이다. 그리고, 그래프를  그 중 1번째 놈에 그리겠다는 의미이다.


#2.그래프를 그리는 작업
plt.plot(x,y)
plt.title("1번 subplot")
plt.grid(True)




#3.그 다음에, 2번째 위치를 잡아야 한다.

plt.subplot(1,2,2) #이 부분을 생략하면, 2번째 그래프가 아예 그려지지 않는다.


#4.그래프 그리기 + 설정
#scatter는 산점도이다.
plt.scatter(x,y)
plt.title("2번 subplot")



#5.그리라고 명령
plt.tight_layout() #axes의 간격을 알아서 조정.
plt.show()
```

```python
#하나의 axes에 여러 그래프를 그리기.

import matplotlib.pyplot as plt
plt.plot(x,y,label='비례')
plt.plot(x,y[::-1],label='반비레') #이거는 반대로 그린다는 의미이다.


plt.legend(loc='center left') #앞서 코딩했던 label을 보여준다. loc을 통해 보여주는 위치를 바꿀 수 있다.
#위치 잡는 것을 별것 아니라고 생각하면 안된다. 그래프가 복잡할 수록 이놈의 위치가 중요할 수 있다.
plt.show()
```

## Figure 와 Axes 객체를 이용해 그리기

- AxesSubplot 객체(Axes를 표현하는 객체)의 메소드들을 이용해 그래프를 그린다.
- axes 생성 방법
    - plt.gca(): AxesSubplot
        - 하나의 figure에 하나의 Axes 만 사용할 경우 사용.
    - 하나의 figure에 여러개의 Axes들을 사용할 경우
        - figure.add_subplot() 메소드 이용
            - figure를 먼저 생성후 axes 들을 추가
        - pyplot.subplots() 함수를 이용
            - figure와 axes배열을 동시에 생성

```python
#아 진짜 변비 진심.....

axes = plt.gca()
print(type(axes))
axes.plot(x,y)


#설정하는 함수들. 근데 이 함수들이 다른 것과는 다르다.
#그 다른 점을 잘 파악하면서 코드를 따져보자.
axes.set_title("제목") #제목설정
axes.set_xticks(ticks=[1,1.5,2,2.5,3,3.5,4,4.5,5])
axes.set_xlabel("X출 라벨") #plt.xlabel("x라벨")
axes.set_ylabel("Y출 라벨")



axes.grid(True)
plt.show()
```

### Figure에 하나의 axes 그리기

### 하나의 Figure에 여러개의 axes 그리기
1. <b style='font-size:1.3em'>figure.add_subplot() 메소드 이용</b>
    - figure객체에 axes를 추가하는 형태
    - nrows(총행수), ncols(총열수), index(axes위치) 지정

```python
#figure 객체 생성


fig = plt.figure()
print(type(fig)) #figure의 data type을 잘 확인할 수 있다.



#1번째 그래프 설정
axes1 = fig.add_subplot(2,2,1) #2x2의 1번 인덱스에~~!
axes1.plot(x,y)
axes1.set_title("1번 axes")



#2번째 그래프 설정
 
axes2 = fig.add_subplot(2,2,3) #2x2의 3번째에 그리는 것이다.
axes2.plot(y,x)
axes2.set_title("2번 axes")


axes3 = fig.add_subplot(1,2,2) #2x2가 아닌 1x2이기 때문에, 그려지는 부분이 다르다.
axes3.bar(['A','B'],[10,20])
axes3.set_title("3번 axes") #set_title을 한다.

plt.tight_layout()
plt.show()
```

2. <b style='font-size:1.3em'>pyplot.subplots()</b>
    - 미리 나눠놓고 하는 것이다.
    - nrows, ncols 로 axes개수와 위치 지정
    - 반환: figure와 axes(subplot)들을 담은 **ndarray**

```python
#또 하나의 중요한 코드이다. 복습을 철저히 하자!!

import matplotlib.pyplot as plt

fig,axes = plt.subplots(2,2,figsize=(10,10)) #출력을 담당하는 것은 axes이다.
print(type(fig))
print(type(axes),axes.shape)

axes[0,0].plot(x,y,c="k")
axes[0,1].plot(y,x,c="#71b079")
axes[1,0].plot(x,y[::-1])
axes[1,1].plot(x,x)

plt.show()
```

## Style
- Style: 그래프의 여러 시각효과들을 미리 설정해 놓은 것
- matplotlib는 다양한 스타일들을 미리 정의해 놓고 있다.
    - [스타일목록](https://matplotlib.org/stable/gallery/style_sheets/style_sheets_reference.html)
    - `plt.style.use()` 함수 이용해 지정
    - `plt.style.available`
        - 사용할 수 있는 스타일 이름 조회

```python
import matplotlib.pyplot as plt
import numpy as np

```

```python
plt.style.use("ggplot")
plt.style.use("dark_background") #style.use을 통해 스타일을 다룰 수 있다.
x = np.linspace(1,10,num=100) #1~100 범위를 100등분했을 때의 분위수 가지는 배열



plt.plot(x,x+3)
plt.plot(x,x+2)
plt.plot(x,x)
plt.plot(x,x-1)
plt.show()
```
